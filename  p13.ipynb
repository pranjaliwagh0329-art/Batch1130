### Measure of Disperssion

```python
import numpy as np
```

```python
ages=[23,26,29,34,22,38,56,84,59.48,37,28]
```

```python
np.var(ages,axis=0)
```

```python
def variance(data):
    n=len(ages)
    mean=sum(data)/n
    deviation=[(x-mean)** 2 for x in data]
    varianve=sum(deviation)/(n-1)
    return variance
```

```python
variance(ages)
```
